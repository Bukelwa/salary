#### l)	Create a scatterplot showing the relationship between Years Worked and Salary (don’t forget to insert a trend line). What is the relationship between Years Worked and Salary?

```python
# here we will extract the years worked column from our salaryDf 

yearsworked = salaryDf['yearsworked']
salary = salaryDf['salary']

salary = salary.drop(salary.index[[208]])
yearsworked = yearsworked.drop(yearsworked.index[[208]])

x = yearsworked
y = salary

# plotting the trendline

z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x), "red")

# plotting a graph
plt.scatter(x,y)
plt.ylabel('Salary')
plt.xlabel('Years worked')
plt.title('Relationship between Years Worked and Salary')


plt.show()
```
